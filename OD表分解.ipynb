```python
import pandas as pd
```

```python
df = pd.read_csv(r"E:\yyz-python\百度迁徙（11.22）\成品\OD表.csv",encoding = "gbk")
```

```python
vv=df[2:]
vv
```

```python
jj = pd.DataFrame()
jj["O"]=df["O"]
jj["D"]=df["D"]
jj["2021-09-15"]=df["2021-09-15"].astype("float")
jj
```

```python
def zhuanOD(mm):
    jj = pd.DataFrame()
    jj["O"]=df["O"]
    jj["D"]=df["D"]
    jj[mm]=df[mm].astype("float")
    oo=jj.pivot_table(index="O",columns="D",values=mm)
    oo=oo.fillna(0)
    oo.to_csv("E:\yyz-python\百度迁徙（11.22）\各日OD\\"+mm+".csv",encoding="gbk")
```

```python
from tqdm import tqdm
for i in tqdm(vv):
    zhuanOD(i)
```
