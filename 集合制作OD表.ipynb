```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
```

```python
path = r"E:\yyz-python\百度迁徙（11.22）\市到市转换后"
reader = os.listdir(path)
```

```python
uu=[]
for sheet in tqdm(reader):
    ll = pd.DataFrame()
    df = pd.read_csv(path + "/" + sheet)
    ll = df.drop(["Unnamed: 0","index","Unnamed: 0.1","Unnamed: 0.1.1","o_city","2021-09-13"],axis=1)
    ll["city_name"] = df["city_name"]+sheet[:-4]
    uu.append(ll)
```

```python
hh = pd.read_csv("E:\yyz-python\百度迁徙（11.22）\市到市转换后\-七台河市.csv")
hh= hh.drop(["Unnamed: 0","index","Unnamed: 0.1","Unnamed: 0.1.1","o_city","2021-09-13"],axis=1)
kk = hh.columns.to_list()
gg = pd.DataFrame(columns = kk)
gg
```

```python
for i in tqdm(uu):
    gg= gg.append(i)
gg
```

```python
gg.to_csv(r"E:\yyz-python\百度迁徙（11.22）\gg.csv",encoding = "gbk")
```
